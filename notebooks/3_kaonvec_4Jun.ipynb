```python
import numpy as np
import awkward as ak
from coffea.nanoevents.methods import vector
ak.behavior.update(vector.behavior)

n = 10
vec = ak.zip(
        {
            "x": np.random.normal(size=n),
            "y": np.random.normal(size=n),
            "z": np.random.normal(size=n),
            "e": np.full(n, 1.),
        },
        with_name="LorentzVector",
    )
print('theta ',vec.theta)
print('theta ',np.arctan2(vec.r, vec.z))
print('r ',np.tan(vec.theta)*vec.z)
print('r ',vec.r)

print('z ',vec.r/(np.tan(vec.theta)))
print('z ',vec.r)
      
print('x^2 + y^2 + z^2 ',vec.x*vec.x + vec.y*vec.y + vec.z*vec.z)
print('rho2 ',vec.rho2)
print('rho ',vec.rho)

print('x^2 + y^2 ',vec.x*vec.x + vec.y*vec.y)
print('r^2 ',vec.r2)
print('r ',vec.r)

print('phi ',np.arctan2(vec.py, vec.px))
print('phi ',vec.phi)

print('rho * sin(theta)',vec.rho*np.sin(vec.theta))
print('r',vec.r)

print('rho * sin(theta) * cos(phi)',vec.rho*np.cos(vec.phi)*np.sin(vec.theta))
print('px',vec.px)

# imagine new_theta
# z = vec.rho * np.cos(new_theta)
# x = vec.rho * np.cos(vec.phi) * np.sin(new_theta)
# y = vec.rho * np.sin(vec.phi) * np.sin(new_theta)      

new_theta = np.full(n, 0.01)
new_energy = np.full(n, 2)

new_vec = ak.zip(
        {
            "x": vec.rho * np.cos(new_theta),
            "y": vec.rho * np.cos(vec.phi) * np.sin(new_theta),
            "z": vec.rho * np.sin(vec.phi) * np.sin(new_theta),
            "e": new_energy,
        },
        with_name="LorentzVector",
    )

# The magnitude (rho in spherical coordinate system).
# tan(theta) = r/z
# tan(phi) = y/x
# cos(theta) = 
#numpy.arctan2(self.r, self.z)
```
